```python
class Lazy:
    def __init__(self, a, b):
        self._a = a
        self._b = b
        self.average = (a + b) / 2
    
    @property
    def a(self):
        return self._a
    
    @a.setter
    def a(self, value):
        self._a = value
        self.average = (self._a + self._b) / 2

# Average before setting a
L = Lazy(1, 2)
print('Average before setting a:', L.average)

# Set a to 3
L.a = 3
print('Average after setting a:', L.average)
```
